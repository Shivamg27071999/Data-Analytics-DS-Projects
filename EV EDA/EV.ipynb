```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Univeriate Analysis

```python
# Plotting frequency of categorical variables

# Plot for 'Make'
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Make', order=df['Make'].value_counts().index)
plt.title('Frequency of Makes')
plt.show()

# Plot for 'Electric Vehicle Type'
plt.figure(figsize=(6,6))
sns.countplot(data=df, x='Electric Vehicle Type')
plt.title('Distribution of Electric Vehicle Type')
plt.show()

# Plot for 'CAFV Eligibility'
plt.figure(figsize=(6,6))
sns.countplot(data=df, x='Clean Alternative Fuel Vehicle (CAFV) Eligibility')
plt.title('CAFV Eligibility Distribution')
plt.show()
```

```python
# Distribution of 'Base MSRP'
plt.figure(figsize=(8,6))
sns.histplot(df['Base MSRP'].dropna(), kde=True, bins=30)
plt.title('Distribution of Base MSRP')
plt.show()

# Distribution of 'Electric Range'
plt.figure(figsize=(8,6))
sns.histplot(df['Electric Range'].dropna(), kde=True, bins=30)
plt.title('Distribution of Electric Range')
plt.show()

# Distribution of 'Model Year'
plt.figure(figsize=(8,6))
sns.histplot(df['Model Year'].dropna(), kde=True, bins=30)
plt.title('Distribution of Model Year')
plt.show()
```

## Bivariate analysis

```python
# Relationship between 'Make' and 'Electric Vehicle Type'
plt.figure(figsize=(12,8))
sns.countplot(data=df, y='Make', hue='Electric Vehicle Type', order=df['Make'].value_counts().index)
plt.title('Electric Vehicle Type by Make')
plt.show()

# Relationship between 'State' and 'CAFV Eligibility'
plt.figure(figsize=(12,8))
sns.countplot(data=df, x='State', hue='Clean Alternative Fuel Vehicle (CAFV) Eligibility')
plt.title('CAFV Eligibility by State')
plt.xticks(rotation=90)
plt.show()
```

```python
# Boxplot of 'Base MSRP' by 'Make'
plt.figure(figsize=(12,8))
sns.boxplot(data=df, y='Make', x='Base MSRP')
plt.title('Base MSRP by Make')
plt.show()

# Boxplot of 'Electric Range' by 'Electric Vehicle Type'
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Electric Vehicle Type', y='Electric Range')
plt.title('Electric Range by Electric Vehicle Type')
plt.show()

# Boxplot of 'Base MSRP' by 'State'
plt.figure(figsize=(12,8))
sns.boxplot(data=df, x='State', y='Base MSRP')
plt.title('Base MSRP by State')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot between 'Base MSRP' and 'Electric Range'
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Base MSRP', y='Electric Range')
plt.title('Scatter Plot of Base MSRP vs Electric Range')
plt.show()

# Scatter plot between 'Model Year' and 'Electric Range'
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Model Year', y='Electric Range')
plt.title('Scatter Plot of Model Year vs Electric Range')
plt.show()

# Correlation matrix of numerical variables
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Boxplot to detect outliers in 'Base MSRP'
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Base MSRP')
plt.title('Boxplot for Base MSRP (Detecting Outliers)')
plt.show()

# Boxplot to detect outliers in 'Electric Range'
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Electric Range')
plt.title('Boxplot for Electric Range (Detecting Outliers)')
plt.show()
```

## Create a Choropleth using plotly.express to display the number of EV vehicles based on location.

```python
# Install plotly if you haven't already
!pip install plotly
import plotly.express as px

# Load your dataset
# Aggregate data by 'State' to count the number of EVs per state
state_ev_count = df.groupby('State').size().reset_index(name='EV Count')

# Display the first few rows to verify the aggregation
print(state_ev_count.head())

# Create a choropleth map using plotly.express
fig = px.choropleth(state_ev_count,
                    locations='State',  # Column containing the state codes
                    locationmode="USA-states",  # Use USA state mode
                    color='EV Count',  # Column with the count of EVs
                    color_continuous_scale='Blues',  # Color scale for the map
                    scope="usa",  # Focus the map on the USA
                    title='Number of EVs by State',
                    labels={'EV Count': 'Number of EVs'})  # Label for the color bar

# Show the map
fig.show()
```

##  Create a Racing Bar Plot to display the animation of EV Make and its count each year.

```python
# Install plotly if you haven't already
!pip install plotly

import pandas as pd
import plotly.express as px



# Check the first few rows to confirm the structure of the dataset
df.head()

# Group the data by 'Make' and 'Model Year' to get the count of EVs produced by each make per year
make_year_count = df.groupby(['Model Year', 'Make']).size().reset_index(name='EV Count')

# Create an animated bar plot
fig = px.bar(make_year_count,
             x='Make',
             y='EV Count',
             color='Make',  # Coloring bars by the 'Make'
             animation_frame='Model Year',  # Animation based on 'Model Year'
             animation_group='Make',  # Ensure the same 'Make' stays connected across frames
             range_y=[0, make_year_count['EV Count'].max() * 1.1],  # Set y-axis range
             title='Racing Bar Plot: EV Make Count by Year',
             labels={'EV Count': 'Number of EVs'})  # Label for y-axis

# Customize the layout (optional)
fig.update_layout(xaxis={'categoryorder': 'total descending'},  # Sort by count
                  transition={'duration': 1000})  # Animation duration (1000ms per year)

# Show the animated plot
fig.show()
```
